#### Formatting the style of your plot


```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20]) # The axis function in the example above takes a list of [xmin, xmax, ymin, ymax] and specifies the viewport of the axes.
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'go--', linewidth=2, markersize=20)
plt.axis([0, 6, 0, 20]) 
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'go--', color='blue', marker='o', linestyle='dashed',linewidth=2, markersize=20)
plt.axis([0, 6, 0, 20]) 
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'go--', color='blue', marker='h', linestyle='dashed',linewidth=2, markersize=20)
plt.axis([0, 6, 0, 20]) 
plt.show()
```

```python
import numpy as np
data = {'a' : np.arange(50),
        'c' : np.random.randint(0, 50, 50),
        'd' : np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
